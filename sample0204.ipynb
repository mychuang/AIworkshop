### 認識自訂類別

```python
class House:
  def __init__(self):
    self.size = 10
    self.numOfBath = 5
    self.numOfDoor = 3

  def openDoor(self):
    print("open Door")
    pass

myHouse = House()

myHouse.openDoor()
```

### 對類別中的屬性進行取值

```python
class House:
  def __init__(self):
    self.size = 10
    self.numOfBath = 5
    self.numOfDoor = 3

  def openDoor(self):
    print("open Door")
    pass
  
  def getSize(self):
    return self.size

myHouse = House()
myHouseSize = myHouse.getSize()
print(myHouseSize)
```

### 傳變數給類別

```python
class House:
  def __init__(self):
    self.__size = 10
    self.__numOfBath = 5
    self.__numOfDoor = 3

  def openDoor(self):
    print("open Door")
    pass
  
  def getSize(self):
    return self.__size
  
  def setSize(self, size):
    self.__size = size
    pass

myHouse = House()
myHouse.setSize(100)
myHouseSize = myHouse.getSize()
print(myHouseSize)

# 請試著改寫成私有屬性(類別封裝)
```

### 使用套件(外部的class)

```python
import numpy

# numpy 是 python 常用的矩陣運算模組
# 可以透過 numpy 來創建陣列或矩陣

a = numpy.array([1,2,3])

print(type(a))

# sum方法，能直接對numpy.ndarray做加總
print(a.sum())
```
