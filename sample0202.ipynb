### 認識Function

```python
# 我們可以利用function幫我們做計算

def add(a, b):
  return a+b

c = add(8, 10)
print(c)
```

```python
# Funtion 內部也能存在不同的function，進行更複雜的工作

def showMyName(name):
  print(name)

showMyName("Miller")
```

```python
def addthreeVar(var1, var2, var3):
  result = add(var2, var3)
  result2 = var1 + result
  return result2

  
print(addthreeVar(5, 10, 15))
```
