
### 準備資料

```python
import numpy as np

# if want [a, b): (b-a) * random_sample() + a
width = np.append(3*np.random.random(100), 3*np.random.random(100)+3)

hight = np.append(3*np.random.random(100)+3, 3*np.random.random(100))

import matplotlib.pyplot as plt
plt.plot(width[:100], hight[:100], 'yo') #for class-1
plt.plot(width[100:], hight[100:], 'bo') #for class-2
plt.xlabel('width', color='#1C2833')
plt.ylabel('hight', color='#1C2833')
plt.show()
```

### define function

```python
# step1: 決定模型 y = wx
# w: weighting
w = 0.25 #一開始先任意猜一個權重值

x0 = np.linspace(0,6,200)
y0 = w*x0

plt.plot(x0, y0, 'r')
plt.plot(width[:100], hight[:100], 'yo') #for class-1
plt.plot(width[100:], hight[100:], 'bo') #for class-2
plt.title("True Data")
plt.xlabel("wide")
plt.ylabel("hight")
plt.show()
```

### 定義Loss

```python
"""
1. Loss = target - y

2. 設定目標值: 
  已知 hight = np.append(3*np.random.random(100)+3, 3*np.random.random(100)) 0~6
  y 不能剛好穿過x, 因為分類線必須能正確區分不同資料

3. 設定學習率: lr
"""
lr = 0.5

target = np.array([])
for i in range(len(width)):
  if(width[i] >= 3):
    target = np.append(target, width[i]-0.1)
  elif (width[i] < 3):
    target = np.append(target, width[i]+0.1)
print(target.shape)
```

### Pick the best function

```python
for i in range(200):
  y = w * width[i]
  loss = target[i] - y
  dw = lr * (loss/width[i])
  w = w + dw
  print("weighting is updated: ", w)
result = w * x0
```

```python
plt.plot(x0, result, 'r')
plt.plot(width[:100], hight[:100], 'yo') #for class-1
plt.plot(width[100:], hight[100:], 'bo') #for class-2
plt.title("True Data")
plt.xlabel('width', color='#1C2833')
plt.ylabel('hight', color='#1C2833')
plt.show()
```
