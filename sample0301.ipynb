### 準備資料

```python
import numpy as np

width = np.array([3, 1])
hight = np.array([1, 3])

print(type(width))
```

```python
import matplotlib.pyplot as plt
plt.plot(width[0], hight[0], 'yo') #for lab1
plt.plot(width[1], hight[1], 'bo') #for lab2
plt.title("True Data")
plt.xlabel("wide")
plt.ylabel("hight")
plt.show()
```

### define function

```python
# step1: 決定模型 y = wx
# w: weighting
w = 0.25 #一開始先任意猜一個權重值

x0 = np.linspace(1,3,3)
y0 = w*x0

plt.plot(x0, y0, 'r')
plt.plot(width[0], hight[0], 'yo') #for lab1
plt.plot(width[1], hight[1], 'bo') #for lab2
plt.title("True Data")
plt.xlabel("wide")
plt.ylabel("hight")
plt.show()
```

### 定義Loss

```python
"""
1. Loss = target - y

2. 設定目標值: 
已知 hight = np.array([1, 3])
#target = np.array([1, 0]) 不能跟height一樣，因為我們不希望分類線剛好穿過資料點
故使用: target = np.array([1.1, 2.9])

3. 設定學習率: lr
"""
target = np.array([1.1, 2.9])
lr = 0.5
```

### Pick the best function

```python
for i in range(2):
  y = w * width[i]
  loss = target[i] - y
  dw = lr * (loss/width[i])
  w = w + dw
  print("weighting is updated: ", w)
result = w * x0
```

```python
plt.plot(x0, result, 'r')
plt.plot(width[0], hight[0], 'yo') #for lab1
plt.plot(width[1], hight[1], 'bo') #for lab2
plt.title("True Data")
plt.xlabel("wide")
plt.ylabel("hight")
plt.show()
```
