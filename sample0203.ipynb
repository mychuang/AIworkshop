### 認識自訂類別

```python
# __init__(): 為python內建函式，代表要初始化此類別
# 由於自訂類別python並不認識，"初始化"代表將自訂類別記錄起來，提供在這一支程式中使用
# __init__(self): self表示此類別描述自己本身
# 在物件導向程式語言中，init()稱為"建構函式"
class House:
  def __init__(self):
    self.size = 10
    self.numOfBath = 5
    self.numOfDoor = 3

myHouse = House() # 動態宣告下，依據等號右邊的值，能產生對應類別的物件

print(type(myHouse))
```

### 對類別中的屬性進行取值

```python
print(myHouse.size)

doors = myHouse.numOfDoor

print(doors)
```

### 傳變數給類別

```python
class House:
  def __init__(self, size, numOfBath, numOfDoor):
    self.size = size
    self.numOfBath = numOfBath
    self.numOfDoor = numOfDoor

myHouse = House(20, 15, 100) # 動態宣告下，依據等號右邊的值，能產生對應類別的物件
print(myHouse.size)
```

### 類別封裝

```python
# 加上雙底線__，不能從外部對該類別屬性作修改，該屬性稱為"私有屬性(private)"，反之稱為"公用屬性(public)"
# 若該類別不提供能修改的屬性，我們稱為"封裝"
class House:
  def __init__(self):
    self.__size = 10
    self.__numOfBath = 5
    self.__numOfDoor = 3

myHouse = House() # 動態宣告下，依據等號右邊的值，能產生對應類別的物件

#print(myHouse.__size)
```
